```python
run_query("""
select e.emp_name,e.emp_id, d.dept_name,d.location
from employees e
inner join departments d
on e.dept_id=d.dept_id limit 10""")
```

```python
run_query("""
select e.emp_name,e.emp_id, d.dept_name,d.location
from employees e
left join departments d
on e.dept_id=d.dept_id limit 10""")
```

```python
run_query("""
select d.dept_name, count(e.emp_id) as headcount,d.budget
from departments d
left join employees e
on d.dept_id=e.dept_id
group by d.dept_name
order by headcount desc
""")
```

```python
run_query("""select emp_name,salary
from employees where  salary >
(select avg(salary) from employees)
order by salary desc""")
```

```python
run_query("""
select emp_id,emp_name,dept_id,salary
from employees where salary in
(select max(salary) from employees
group by dept_id)
order by salary desc""")
```

```python
run_query("""
select emp_name,salary,
rank() over (order by salary desc) as salary_rank
from employees
limit 10
""")
```

```python
run_query("""
select emp_id,emp_name,gender,salary from employees group by salary
order by salary desc limit 10""")
```

```python
run_query("""
select emp_name, emp_id, salary,gender,
rank() over (order by salary desc) as highest_salary_per_dept
from employees
group by dept_id
limit 10""")
```

```python
run_query("""
select emp_name, dept_id,salary,
rank() over (partition by dept_id order by salary desc) as dept_rank
from employees
order by dept_id, dept_rank
limit 20
""")
```

```python
run_query("""
SELECT emp_name, dept_id, salary,
RANK() OVER (PARTITION BY dept_id ORDER BY salary DESC) AS dept_rank
FROM employees
ORDER BY dept_id, dept_rank
LIMIT 20
""")
```

```python
run_query ("""
    select emp_name,salary,
sum(salary) over (order by emp_id rows between unbounded preceding and current row ) as running_total
from employees
limit 10
""")
```
